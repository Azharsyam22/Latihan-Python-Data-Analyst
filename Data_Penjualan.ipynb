```python
import pandas as pd
import matplotlib.pyplot as plt

data_transaksi = {
    'Tanggal': ['2024-12-01', '2024-12-01', '2024-12-02', '2024-12-02', '2024-12-03', '2024-12-03'],
    'Nama_Barang': ['Laptop Gaming Asus', 'Mouse Logitech', 'Mechanical Keyboard', 'Monitor Samsung', 'Laptop Gaming Asus', 'Mouse Logitech'],
    'Kategori': ['Laptop', 'Aksesoris', 'Aksesoris', 'Monitor', 'Laptop', 'Aksesoris'],
    'Harga_Satuan': [15000000, 150000, 800000, 2500000, 15000000, 150000],
    'Jumlah_Terjual': [2, 10, 5, 4, 1, 20]
}


df = pd.DataFrame(data_transaksi)

df['Total_Omzet'] = df['Harga_Satuan'] * df['Jumlah_Terjual']

print("=== DATA TRANSAKSI LENGKAP ===")
print(df)
print("\n")

laporan_penjualan = df.groupby('Nama_Barang')['Total_Omzet'].sum().sort_values(ascending=False)

print("=== LAPORAN OMZET PER BARANG (Tertinggi ke Terendah) ===")
print(laporan_penjualan)

plt.figure(figsize=(10, 6))

plt.bar(laporan_penjualan.index, laporan_penjualan.values, color=['#4CAF50', '#FF9800', '#2196F3', '#F44336'])

plt.title('Total Pendapatan per Produk')
plt.xlabel('Nama Produk')
plt.ylabel('Total Rupiah (dalam Puluhan Juta)')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.5)


plt.ticklabel_format(style='plain', axis='y')

plt.show()
```
