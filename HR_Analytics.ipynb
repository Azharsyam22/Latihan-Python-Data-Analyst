```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = {
    'Nama': [f'Dev_{i}' for i in range(1, 21)],
    'Pengalaman_Tahun': [1, 2, 3, 4, 5, 2, 8, 3, 6, 9, 1, 10, 4, 7, 5, 2, 12, 3, 15, 1],
    'Gaji_Juta': [5, 6, 7.5, 9, 11, 5.5, 18, 7, 13, 20, 4.5, 22, 9.5, 15, 10, 6, 25, 7.2, 30, 5],
    'Role': ['Junior', 'Junior', 'Mid', 'Mid', 'Senior', 'Junior', 'Lead', 'Mid', 'Senior', 'Lead',
             'Junior', 'Manager', 'Mid', 'Senior', 'Senior', 'Junior', 'Manager', 'Mid', 'CTO', 'Junior']
}

df_mentah = pd.DataFrame(data)


df_mentah.to_csv('Data_Gaji_Programmer.csv', index=False)
print("File CSV berhasil dibuat! Cek folder di kiri.")
```

```python
df = pd.read_csv('Data_Gaji_Programmer.csv')

print("=== DATA 5 KARYAWAN PERTAMA ===")
print(df.head())
print("\n")


print(f"Gaji Tertinggi: {df['Gaji_Juta'].max()} Juta")
print(f"Gaji Terendah: {df['Gaji_Juta'].min()} Juta")
print(f"Rata-rata Pengalaman: {df['Pengalaman_Tahun'].mean():.1f} Tahun")

plt.figure(figsize=(10, 6))

scatter = plt.scatter(df['Pengalaman_Tahun'], df['Gaji_Juta'], c=df['Gaji_Juta'], cmap='viridis', s=100, alpha=0.8)

plt.colorbar(scatter, label='Tingkat Gaji')
plt.title('Hubungan Pengalaman Kerja vs Gaji Programmer')
plt.xlabel('Lama Pengalaman (Tahun)')
plt.ylabel('Gaji (Juta Rupiah)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
